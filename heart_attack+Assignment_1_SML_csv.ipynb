```
import numpy as np
import matplotlib.pyplot as plt

X = np.array([2.3, 2.5, 3.6, 2.8, 3.1, 2.9, 3.2, 2.7, 2.8, 3.0])

bins = 5

hist, bin_edges = np.histogram(X, bins=bins, density=True)

plt.hist(X, bins=bin_edges, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title("Histogram with Density")
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.show()

print("Density for each bin:", hist)
print("Bin edges:", bin_edges)
```

```
import numpy as np
from scipy.stats import skew, kurtosis, iqr

dependent = np.array([4, 5, 8, 2, 4, 2, 5])
independent = np.array([5, 6, 3, 8, 3, 7, 8])

def calculate_statistics(data):
    stats = {
        'Mean': np.mean(data),
        'Median': np.median(data),
        'Mode': np.bincount(data).argmax(),
        'Variance': np.var(data, ddof=1),
        'Standard Deviation': np.std(data, ddof=1),
        'Range': np.ptp(data),
        'IQR': iqr(data),
        'Skewness': skew(data),
        'Kurtosis': kurtosis(data)
    }
    return stats

def print_statistics(label, stats):
    print(f"{label} Statistics:")
    for key, value in stats.items():
        print(f"{key}: {value}")
    print()

dependent_stats = calculate_statistics(dependent)
independent_stats = calculate_statistics(independent)

print_statistics("Dependent Variable", dependent_stats)
print_statistics("Independent Variable", independent_stats)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
from scipy.stats import skew, kurtosis, iqr
data = pd.read_csv('/content/drive/MyDrive/heart_attack_dataset.csv', usecols=['Age'])
print(data.tail())
print(data.head(20))
bins = 100
hist, bin_edges = np.histogram(data['Age'], bins=bins)

colors = []
for frequency in hist:
    if frequency < 10:
        colors.append('green')
    elif 10 <= frequency < 20:
        colors.append('blue')
    elif 20 <= frequency < 30:
        colors.append('purple')
    else:
        colors.append('red')
for i in range(len(hist)):
    plt.bar(bin_edges[i], hist[i], width=bin_edges[i+1]-bin_edges[i], color=colors[i], edgecolor='black')

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
mean_age = np.mean(data['Age'])
median_age = np.median(data['Age'])
mode_age = stat.mode(data['Age'])
std_dev_age = np.std(data['Age'])
skewness_age = skew(data['Age'], axis=0, bias=True)
kurtosis_age = kurtosis(data['Age'], axis=0, bias=True)
iqr_age = iqr(data['Age'])
range_age = np.ptp(data['Age'])
print("The mean of the data =", mean_age)
print("The median of the data =", median_age)
print("The mode of the data =", mode_age)
print("The standard deviation of the data =", std_dev_age)
print("The skewness of the data =", skewness_age)
print("The kurtosis of the data =", kurtosis_age)
print("The interquartile range (IQR) of the data =", iqr_age)
print("The range of the data =", range_age)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
from scipy.stats import skew, kurtosis, iqr

data = pd.read_csv('/content/drive/MyDrive/heart_attack_dataset.csv', usecols=['Cholesterol (mg/dL)', 'Blood Pressure (mmHg)'])

print(data.tail())
print(data.head(20))

bins = 100

hist_chol, bin_edges_chol = np.histogram(data['Cholesterol (mg/dL)'], bins=bins)
hist_bp, bin_edges_bp = np.histogram(data['Blood Pressure (mmHg)'], bins=bins)

plt.figure(figsize=(14, 7))

plt.bar(bin_edges_chol[:-1], hist_chol, width=bin_edges_chol[1] - bin_edges_chol[0], color='blue', alpha=0.7, label='Cholesterol', edgecolor='black')

plt.bar(bin_edges_bp[:-1], hist_bp, width=bin_edges_bp[1] - bin_edges_bp[0], color='green', alpha=0.7, label='Blood Pressure', edgecolor='black')

plt.title('Cholesterol and Blood Pressure Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()

mean_chol = np.mean(data['Cholesterol (mg/dL)'])
median_chol = np.median(data['Cholesterol (mg/dL)'])
mode_chol = stat.mode(data['Cholesterol (mg/dL)'])
std_dev_chol = np.std(data['Cholesterol (mg/dL)'])
skewness_chol = skew(data['Cholesterol (mg/dL)'], axis=0, bias=True)
kurtosis_chol = kurtosis(data['Cholesterol (mg/dL)'], axis=0, bias=True)
iqr_chol = iqr(data['Cholesterol (mg/dL)'])
range_chol = np.ptp(data['Cholesterol (mg/dL)'])

mean_bp = np.mean(data['Blood Pressure (mmHg)'])
median_bp = np.median(data['Blood Pressure (mmHg)'])
mode_bp = stat.mode(data['Blood Pressure (mmHg)'])
std_dev_bp = np.std(data['Blood Pressure (mmHg)'])
skewness_bp = skew(data['Blood Pressure (mmHg)'], axis=0, bias=True)
kurtosis_bp = kurtosis(data['Blood Pressure (mmHg)'], axis=0, bias=True)
iqr_bp = iqr(data['Blood Pressure (mmHg)'])
range_bp = np.ptp(data['Blood Pressure (mmHg)'])

print("\nCholesterol Statistics:")
print("The mean =", mean_chol)
print("The median =", median_chol)
print("The mode =", mode_chol)
print("The standard deviation =", std_dev_chol)
print("The skewness =", skewness_chol)
print("The kurtosis =", kurtosis_chol)
print("The interquartile range (IQR) =", iqr_chol)
print("The range =", range_chol)

print("\nBlood Pressure Statistics:")
print("The mean =", mean_bp)
print("The median =", median_bp)
print("The mode =", mode_bp)
print("The standard deviation =", std_dev_bp)
print("The skewness =", skewness_bp)
print("The kurtosis =", kurtosis_bp)
print("The interquartile range (IQR) =", iqr_bp)
print("The range =", range_bp)



```
