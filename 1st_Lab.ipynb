```
import matplotlib.pyplot as plt
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
number_of_students = [0,1,1,2,2,0,2,3,4,3,0,1]
colors = ['green' if n==1 else 'blue' if n==2 else 'purple' if n==3 else 'red' if n==4 else 'grey' for n in number_of_students]
plt.figure(figsize=(10,6))
plt.title ("Birth Month of 19 Students of Batch 37", fontstyle= 'italic')
plt.bar(x=month, height=number_of_students, color=colors)
plt.xlabel("Months")
plt.ylabel("Number of students")
plt.show()
```

```
import matplotlib.pyplot as plt
birth_month = {
    "Jan" : 0,
    "Feb" : 1,
    "Mar" : 1,
    "Apr" : 2,
    "May" : 2,
    "Jun" : 0,
    "Jul" : 2,
    "Aug" : 3,
    "Sep" : 4,
    "Oct" : 3,
    "Nov" : 0,
    "Dec" : 1
}
months = list(birth_month.keys())
counts = list(birth_month.values())
plt.figure(figsize=(10,6))
plt.bar(months, counts, color='skyblue')
plt.xlabel('months')
plt.ylabel('no. of births')
plt.title('no. of births per months')
plt.show()
```
